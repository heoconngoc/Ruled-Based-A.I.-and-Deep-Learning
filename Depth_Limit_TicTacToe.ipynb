```python
# =========================================================
# TITLE: Depth-Limited Minimax
# =========================================================

"""
Minimax with depth limit (no heuristic).
If depth = 0 -> treat as draw (0)
"""

import time
import random

EMPTY=" "
X="X"
O="O"

def create_board():
    return [EMPTY]*9

def get_moves(b):
    return [i for i in range(9) if b[i]==EMPTY]

def check_winner(b):
    wins=[(0,1,2),(3,4,5),(6,7,8),(0,3,6),(1,4,7),(2,5,8),(0,4,8),(2,4,6)]
    for i,j,k in wins:
        if b[i] != EMPTY and b[i] == b[j] == b[k]:
            return b[i]
    return None

def switch(p):
    return O if p==X else X
```

```python
# ---------------------------
def minimax_depth(board, player, depth):
    winner=check_winner(board)

    if winner==X:
        return 1,None
    if winner==O:
        return -1,None
    if not get_moves(board):
        return 0,None
    if depth==0:
        return 0,None

    if player==X:
        best=-999; move=None
        for m in get_moves(board):
            b=board.copy()
            b[m]=X
            val,_=minimax_depth(b,O,depth-1)
            if val>best:
                best=val; move=m
        return best,move
    else:
        best=999; move=None
        for m in get_moves(board):
            b=board.copy()
            b[m]=O
            val,_=minimax_depth(b,X,depth-1)
            if val<best:
                best=val; move=m
        return best,move


def minmax_agent(board, player):
    _,m = minimax_depth(board,player,3)
    return m
```

```python
# ---------------------------
# GAME ENGINE
# ---------------------------
def play_game(agent1, agent2, verbose=False):
    board = create_board()
    player = X

    while True:

        if player == X:
            move = agent1(board, X)
        else:
            move = agent2(board, O)

        board[move] = player

        w = check_winner(board)
        if w:
            return 1 if w == X else -1

        if not get_moves(board):
            return 0

        player = switch(player)
```

```python
def evaluate(agent1, agent2, games=100):

    results = []

    start_time = time.time()

    for i in range(games):

        # swap first move for fairness
        if i % 2 == 0:
            res = play_game(agent1, agent2)
        else:
            res = play_game(agent2, agent1)
            res = -res

        results.append(res)

    end_time = time.time()


    print("\n======================")
    print("RESULTS")
    print("======================")

    print("Agent1 Wins :", results.count(1))
    print("Agent2 Wins :", results.count(-1))
    print("Ties        :", results.count(0))

    print("\nTime taken  :", round(end_time - start_time, 4), "sec")
    print("Avg time/game:", round((end_time - start_time)/games, 6), "sec")
```

```python
# Random agent
def random_agent(board, player):
    return random.choice(get_moves(board))

# Rule-based 1-step agent
def rule_1_step(board, player):

    opponent = switch(player)

    # win
    for m in get_moves(board):
        b = board.copy()
        b[m] = player
        if check_winner(b) == player:
            return m

    # block
    for m in get_moves(board):
        b = board.copy()
        b[m] = opponent
        if check_winner(b) == opponent:
            return m

    return random.choice(get_moves(board))
```

```python
# Minmax agent vs Random_agent
evaluate(minmax_agent, random_agent, games=100)
```

```python
# Minmax agent vs Rule_1_step
evaluate(minmax_agent, rule_1_step, games=100)
```

```python
# Minmax agent vs Minmax agent
evaluate(minmax_agent, minmax_agent, games=100)
```

Depth-limited minimax reduces computation time significantly but may produce suboptimal decisions due to incomplete search.
